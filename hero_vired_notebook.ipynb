A company collects website visitor logs for one day.
Some visitors refresh the page, so their IDs appear multiple times:

visitors = ["U101", "U102", "U101", "U105", "U103", "U105"]


The marketing team also has a list of users they targeted with ads:

targeted = {"U101", "U103", "U110", "U120"}


Find how many targeted users actually visited the website.

Use sets to solve it.

```python
visitors = ["U101", "U102", "U101", "U105", "U103", "U105"]
visitors
```

```python
targeted = {"U101", "U103", "U110", "U120"}
targeted
```

```python
visitors = ["U101", "U102", "U101", "U105", "U103", "U105"]
targeted = {"U101", "U103", "U110", "U120"}

# 1. Convert the visitors list to a set to remove duplicates
unique_visitors = set(visitors)

# 2. Find the intersection (users present in BOTH sets)
successful_targets = unique_visitors.intersection(targeted)

# 3. Count the results
count = len(successful_targets)

print(f"Number of targeted users who visited: {count}")
print(f"The User IDs are: {successful_targets}")
```

```python
visitors = ["U101", "U102", "U101", "U105", "U103", "U105"]
targeted = {"U101", "U103", "U110", "U120"}

# Convert visitors list to a set to remove duplicates
visitor_set = set(visitors)

# Find targeted users who actually visited
visited_targeted = visitor_set & targeted   # intersection

print("Targeted users who visited:", visited_targeted)
print("Count:", len(visited_targeted))
```

```python
# while -> condition based scenario, unaware about how many times execution will happen
# while condition:
#     body statements

correct_pin = '1234'
user_pin = input('Enter your pin')

while user_pin != correct_pin:
    print('Please enter correct pin')
    user_pin = input('Enter correct pin pls')

correct_pin = '1234'
user_pin = input('Enter your pin')
count = 0
while user_pin != correct_pin:
    count+=1
    print('Please enter correct pin')
    user_pin = input('Enter correct pin pls')

print(f'After completing {count} times, you got correct pin')
```
